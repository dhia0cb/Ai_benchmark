```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
print("imports ok")
```

```python
df = pd.read_csv(r'C:\Users\DHIA\benchmark_ai\resultats.csv')
print(df.head())
print(df.describe())
```

```python
best = df.groupby(['tache','outil'])['note_globale'].mean()
print(best)
```

```python
global_score = df.groupby('outil')['note_globale'].mean().sort_values(ascending=False)
print(global_score)
```

```python
plt.figure(figsize=(10,6))
sns.barplot(data=df, x='outil', y='note_globale', hue='tache')
plt.title('Comparaison des outils IA par tâche')
plt.tight_layout()
plt.savefig('../data/graphique.png')
plt.show()
```
